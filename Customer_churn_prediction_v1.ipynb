```python
##########################################################################
# Функции для Downsampling и тестирования параметров
##########################################################################


def downsample(features, target, fraction, random_state):    
    """Randomly deletes fraction of 0-class data.
    
    uses 'sample' and 'shuffle' methods
    returns downsampled results
    """
    
    # Split dataframes by target values (0 and 1)
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    # Create smaller samples for splits with target==0
    # Concat with the rest of data
    features_downsampled = pd.concat(
        [features_zeros.sample(
            frac=fraction, random_state=random_id)]
        + [features_ones])
    
    target_downsampled = pd.concat(
        [target_zeros.sample(
            frac=fraction, random_state=random_id)] 
        + [target_ones])
    
    # Randomly shuffle data
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, 
        target_downsampled, random_state=random_id)
    
    return features_downsampled, target_downsampled


def testing_downsample_balance(X, y, frac_range, random_state):
    """Prints class balances with different frac parameters"""
    
    print("Баланс классов в выборке после downsampling:")
    
    # Save initial balance in a dataset
    balance = y.value_counts(normalize=True).rename('init_balance').to_frame()  
    
    # Calculate and print new balances based on the frac parameter
    for frac in frac_range:
        X_downsampled, y_downsampled = downsample(X, y, frac, random_state)           
        balance = balance.join(
            y_downsampled
            .value_counts(normalize=True)
            .rename('sample_'+str(frac.round(2)))
        )
       
    display(balance)  


def print_downsample_results(model, frac, X, y, 
                             X_valid, y_valid, random_state):    
    """Prints f1-score on validation set after downsampling"""  
    
    # Train model on the downsampled dataset
    X_downsampled, y_downsampled = downsample(X, y, frac, random_state)        
    model.fit(X_downsampled, y_downsampled)
        
    # Calculate f1 scores on train and validation datasets
    train_score = f1_score(y, model.predict(X))  
    valid_score = f1_score(y_valid, model.predict(X_valid))  
    print(f"Frac = {frac:.2f} | F1_score: "
          + f"Обучающая выборка = {train_score:.3f}" 
          + f" | Валидационная выборка = {valid_score:.3f}")

    
def get_downsample_results(model, frac, X, y, 
                           X_valid, y_valid, random_state):    
    '''Performs downsample and returns classification scores.
    
    returns f1_score, recall and precision
    ''' 
    
    # Train model on the downsampled dataset
    X_downsampled, y_downsampled = downsample(X, y, frac, random_state)        
    model.fit(X_downsampled, y_downsampled)
     
    # Make predictions on validation dataset  
    predictions = model.predict(X_valid)
    
    # Save probabilities for roc-auc score
    predictions_proba = model.predict_proba(X_valid)[:, 1]
    
    # Calculate metrics
    f1 = f1_score(y_valid, predictions) 
    recall = recall_score(y_valid, predictions) 
    precision = precision_score(y_valid, predictions) 
    roc_auc = roc_auc_score(y_valid, predictions_proba) 
    
    return f1, recall, precision, roc_auc 


##########################################################################
# Функции для Upsampling и тестирования параметров
##########################################################################

def upsample(features, target, repeat, random_state):    
    """Randomly duplicates 1-class data.
    
    uses 'repeat' as multiplicator and 'shuffle' method
    returns upsampled results
    """
   
    # Split dataframes by target values (0 and 1)
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    # Create duplicate samples for splits with target==1
    # Concat with the rest of data
    features_upsampled = pd.concat(
        [features_zeros] + [features_ones] * repeat)
    
    target_upsampled = pd.concat(
        [target_zeros] + [target_ones] * repeat)
   
    # Randomly shuffle data
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, 
        target_upsampled, random_state=random_id)
    
    return features_upsampled, target_upsampled


def testing_upsample_balance(X, y, repeat_range, random_state):
    """Prints class balances with different parameters repeat_range"""
    
    print("Баланс классов в выборке после upsampling\n"
          + "(для разных параметров 'repeat')")
    
    # Save initial balance in a dataset
    balance = y.value_counts(normalize=True).rename('init_balance').to_frame()  
    
    # Calculate and print new balances based on the frac parameter
    for repeat in repeat_range:
        X_upsampled, y_upsampled = upsample(X, y, repeat, random_state)           
        balance = balance.join(
            y_upsampled
            .value_counts(normalize=True)
            .rename('sample_'+str(repeat))
        )
       
    display(balance)    


def print_upsample_results(model, frac, X, y, 
                           X_valid, y_valid, random_state):    
    """Prints f1-score on validation set after upsampling"""  
    
    # Train model on the upsampled dataset
    X_upsampled, y_upsampled = upsample(X, y, frac, random_state)        
    model.fit(X_upsampled, y_upsampled)
        
    # Calculate f1 scores on train and validation datasets
    train_score = f1_score(y, model.predict(X))  
    valid_score = f1_score(y_valid, model.predict(X_valid))  
    print(f"Frac = {frac:.2f} | F1_score: "
          + f"Обучающая выборка = {train_score:.3f}" 
          + f" | Валидационная выборка = {valid_score:.3f}")

    
def get_upsample_results(model, frac, X, y, 
                         X_valid, y_valid, random_state):    
    '''Performs upsample and returns classification scores.
    
    returns f1_score, recall and precision
    ''' 
    
    # Train model on the upsampled dataset
    X_upsampled, y_upsampled = upsample(X, y, frac, random_state)        
    model.fit(X_upsampled, y_upsampled)
     
    # Make predictions on validation dataset  
    predictions = model.predict(X_valid)
    
    # Save probabilities for roc-auc score
    predictions_proba = model.predict_proba(X_valid)[:, 1]
    
    # Calculate metrics
    f1 = f1_score(y_valid, predictions) 
    recall = recall_score(y_valid, predictions) 
    precision = precision_score(y_valid, predictions) 
    roc_auc = roc_auc_score(y_valid, predictions_proba) 
    
    return f1, recall, precision, roc_auc 


##########################################################################
# Валидационные и обучающие кривые
##########################################################################

def plot_learning_curve(title, model, features, 
                        target, score_name, train_sizes,
                        y_min, y_max):
    
    '''Trains model on cross-validation and plots validation curve 
    
    uses sklearn.model_selection.learning_curve function
    '''

    train_sizes, train_scores, validation_scores = learning_curve(
        estimator=model,
        X=features,
        y=target,
        scoring=score_name,
        train_sizes=train_sizes,
        shuffle=True,
        cv=5,
    )

    # Calculate mean of cross-validation scores 
    train_scores_mean = train_scores.mean(axis=1)
    validation_scores_mean = validation_scores.mean(axis=1)

    plt.style.use("seaborn")
    plt.plot(train_sizes, train_scores_mean, label="Training error")
    plt.plot(train_sizes, validation_scores_mean, label="Validation error")
    plt.ylabel(score_name, fontsize=14)
    plt.xlabel("Training set size", fontsize=14)
    plt.title(title, fontsize=14, y=1.03)
    plt.legend()
    
    # Set y limits for better visualization
    plt.ylim(y_min, y_max)

    plt.show()   
    

def plot_validation_curve(title, model, 
                          features, target, score, 
                          param, param_range, y_min, y_max):

    """Trains model on cross-validation and plots validation curve

    uses sklearn.model_selection.learning_curve function
    """
  
    train_scores, validation_scores = validation_curve(
        estimator=model,
        X=features,
        y=target,
        scoring=score,
        param_name=param,
        param_range=param_range,
    )

    # Calculate mean of cross-validation scores
    train_scores_mean = train_scores.mean(axis=1)
    validation_scores_mean = validation_scores.mean(axis=1)

    plt.style.use("seaborn")
    plt.plot(param_range, train_scores_mean, label="Training error")
    plt.plot(param_range, validation_scores_mean, 
             label="Сross-validation error")
    plt.ylabel(score, fontsize=14)
    plt.xlabel(f"Диапазон значений {param}", fontsize=14)
    plt.title(title, fontsize=18, y=1.03)
    plt.legend()

    # Set y limits for better visualization
    plt.ylim(y_min, y_max)

    plt.show()
    
    
##########################################################################
# Графики корреляций
##########################################################################    
    
def print_features_corr_matrix(df, title, 
                               method_name='pearson', round_decimals=2):
    
    ''' Prints correlation matrix with Pearson coefficients. '''
    
    # Scale font size
    sns.set(font_scale=1.4) 
    
    # Plot correlation  matrix
    # with rounded coefficients
    plt.figure(figsize=(15, 6))
    ax = plt.axes()    
    sns.heatmap(df.corr(method=method_name).round(round_decimals), 
                annot=True, ax=ax)
    
    # Rotate x-labels (for long names)
    plt.xticks(rotation=45)
    
    # Add title
    ax.set_title(title, fontsize=16, y=1.03)

    
def print_target_correlation(df, target_col_name, sort_method="Spearman"):
    """Таблица корреляции с коэффициентами Spearman и Pearson"""

    display(
        df
        # Создаем столбец с корреляцией Спирмена
        .corr(method="spearman")[target_col_name]
        .drop([target_col_name])
        .rename("Spearman")
        .to_frame()
        # Добавляем столбец с корреляцией Пирсона
        # (фокус на линейные зависимости)
        .join(
            df.corr()[target_col_name]
            .drop([target_col_name])
            .rename("Pearson")
            .to_frame()
        )
        # Сортируем по коэффициентам Спирмена
        .sort_values(by=sort_method)
        .style.background_gradient(cmap="coolwarm")
    )        
    
    
##########################################################################
# Тестирование результатов моделей
##########################################################################
    
def add_results_classification(model, X, y, cv=5):
    """Trains model and saves cross-validation results to a dataframe

    Parameters:
            model(estimator object with fit and predict)
            X (array-like of shape): features
            y (array-like of shape): target
            cv (int): number of folds in cross-validation
    Returns:
            results (list): four average  metrics on cross-validation
                f1 score, recall, precision and roc-auc 
    """  
  
    # Calculate average of metrics for cv-folds
    results = [
        cross_val_score(model, X, y, cv=cv, scoring='f1').mean(),
        cross_val_score(model, X, y, cv=cv, scoring='recall').mean(),
        cross_val_score(model, X, y, cv=cv, scoring='precision').mean(),
        cross_val_score(model, X, y, cv=cv, scoring='roc_auc').mean(),
    ]
    
       
    return results  


def print_test_results_classification(models_dict, 
                                      train_features, train_target,
                                      test_features, test_target):
   
    '''Trains selected models and shows results on test
    
    the output contains f1-score and roc-auc score
    '''
    
    for model_name in models_dict:
        
        # Models_dict[0] contains the model
        # Models_dict[1] contains its parameters
        model = models_dict[model_name][0]
        
        # Train and make predictions on test
        model.fit(train_features, train_target)
        predictions = model.predict(test_features)
        
        # Save probabilities for roc-auc score
        predictions_proba = model.predict_proba(test_features)[:, 1]            
        
        # Calculate metrics
        f1 = f1_score(test_target, predictions)
        roc_auc = roc_auc_score(test_target, predictions_proba)
        
        # Print results
        print(f"F1_score: {f1 : .3f} |  roc_auc_score: {roc_auc : .3f} |"
              + f" {model_name}  Параметры: {models_dict[model_name][1]}")

        
def search_best_threshold(trained_model, X_valid, y_valid):
    """
    Shows results of classificator with different thresholds.
    
    Output contains f1, precision and recall scores.
    """
   
    # Take different thresholds and calculate model results   
    for threshold in np.arange(0, 0.9, 0.1):

        # Calculate predictions using pretrained model        
        predicted_valid = model.predict_proba(X_valid)         
        # Take not class results, but their probabilities
        predicted_one_valid = predicted_valid[:, 1]

        # если вероятность > трэшхолда, модель предсказывает 1-й класс
        # если <= трешхолда, 0-й класс
        predicted_valid = np.where(predicted_one_valid > threshold, 1, 0)
     
        # Calculate metrics
        precision = precision_score(y_valid, predicted_valid)
        recall = recall_score(y_valid, predicted_valid)
        f1 = f1_score(y_valid, predicted_valid)
        roc_auc = roc_auc_score(y_valid, predicted_one_valid)

        # Print results
        print(f"Threshold = {threshold:.2f} | F1_score = {f1:.3f} | "
              + f"Precision = {precision:.3f}, Recall = {recall:.3f}| "
              + f"ROC_auc = {roc_auc:.3f}")
        
        
##########################################################################
# ROC-AUC и PR-кривые
##########################################################################

def draw_ROC_AUC_curves(title,
                        valid_features, valid_target,
                        models_with_labels):                         
    """
    Draw ROC-AUC curves for the pretrained models.
    
    Parameters:
        title: plot title
        models_with_labels: dictionary with models
        valid_features, valid_target: validation dataset
    """
    
    # List to save predicted probabilities of each model
    models_probas = []

    # For each model in a dictionary 
    for model_label, model in models_with_labels.items(): 
        
        # Calculate predicted (!)probabilities for "1" class
        models_probas.append(model.predict_proba(valid_features)[:, 1])    
    
    # Plot ROC-AUC curves for all models
    # using results on validation dataset
    plt.figure(figsize=(9, 9))
    for  model_label, model_proba in zip (list(models_with_labels.keys()), 
                                          models_probas): 
        # Calculate FPR and TPR values for each threshold
        fpr, tpr, thresholds = roc_curve(valid_target, model_proba)     
        
        # Calculate roc-auc score  for label
        roc_auc = roc_auc_score(valid_target, model_proba)
        
        # Plot ROC_AUC-curve
        plt.plot(fpr, tpr, label=f'{model_label} (AUC = {roc_auc:.2f})')   

    # Add Base-Line to plot (roc-auc = 0.5)    
    plt.plot([0, 1], [0, 1], 
             linestyle='--', label='Модель NO-SKILL (AUC = 0.5)')
 
    # Limit FPR and TPR values to [0,1]
    plt.ylim([0.0, 1.05]) 
    plt.xlim([0.0, 1.05])

    # Add labels and legend   
    plt.xlabel('False Positive Rate', size=14)
    plt.ylabel('True Positive Rate', size=14)
    plt.legend(fontsize=14)
    plt.title(title, size=16, y=1.02)

    plt.show()  

    
def draw_PR_curves(title,
                   valid_features, valid_target,
                   models_with_labels):                         
    """
    Draw Precision-Recall curves for the pretrained models.
    
    Parameters:
        title: plot title
        models_with_labels: dictionary with models
        valid_features, valid_target: validation dataset
    """
    
    # List to save predicted probabilities of each model
    models_probas = []

    # For each model in a dictionary 
    for model_label, model in models_with_labels.items(): 
        
        # Calculate predicted (!)probabilities for "1" class
        models_probas.append(model.predict_proba(valid_features)[:, 1])    
    
    # Plot PR-curves for all models
    # using results on validation dataset
    plt.figure(figsize=(9, 9))
    for  model_label, model_proba in zip (list(models_with_labels.keys()), 
                                          models_probas): 
        # Calculate Precision and Recall values for each threshold
        precision, recall, thresholds = precision_recall_curve(
            valid_target, model_proba)
        
        # Calculate auc-score (average precision) for label        
        auc_score = auc(recall, precision)
        
        # Plot ROC_AUC-curve
        plt.step(recall, precision, where='post', 
                 label=f'{model_label} (AUC = {auc_score:.2f})')

    # Add Base-Line to plot as a horizontal line
    # proportion of "1" class
    baseline_model = sum(valid_target == 1) / len(valid_target) 
    plt.plot([0, 1], 
             [baseline_model, baseline_model], 
             linestyle='--', label='Модель NO-SKILL (AUC = 0.2)')

    # Limit Precision and Recall values to [0,1]
    plt.ylim([0.0, 1.05]) 
    plt.xlim([0.0, 1.05])

    # Add labels and legend   
    plt.xlabel('Recall', size=14)
    plt.ylabel('Precision', size=14)
    plt.legend(fontsize=14)
    plt.title(title, size=16, y=1.02)

    plt.show()
```

#### Категории. Encoding и влияние на корреляцию

В п.3.1 в матрицу корреляции не попали **две категории**, которые могут оказаться важными для предсказательной модели:

- `geography` - Испания, Франция, Германия
- `gender` - Ж и М

1-й вариант - сделать аналог *OrdinalEncoding* с помощью map

2-й вариант - *One-Hot-Encoding*


План-кабан:

1. Посмотреть, как будет **меняться корреляция** в зависимости от типа кодирования


2. Понять, в чем заключается суть "разрекламированной" `dummy-ловушки` в OHE


3. Выбрать `подходящие методы` и взять на заметку для будущих проектов

```python
# Функция one-hot encoding для пайплайна
def ohe_all(df):
    df = pd.get_dummies(df) 
    return df


print('Корреляция категорий, One-Hot-Encoding:')
display(df_ml[['gender', 'geography', 'exited']] 
        .pipe(ohe_all)
        .corr(method='spearman')['exited']
        .drop(['exited']).to_frame()
        .sort_values(by='exited')
        .style.background_gradient(cmap='coolwarm'))
```

**Вывод 1**

1. У `gender_Male` и `gender_Female` одинаковые корреляции. Это отличный пример той самой **дамми-ловушки**.
В `Gender` есть только два значения, поэтому OHE-encoding по сути задублировал нам одни и те же данные.


2. Для `Gender` будем использовать классический **mapping** (0 и 1).


С **географией** все интереснее, сравним значения корреляции с `долей ушедших клиентов` в разных странах.

**Вывод 2**

1. Вес **Германии** ощутимый. Немцев у банка всего 25%, и целых `30%` из них - `ушедшие клиенты`.


2. С учетом одинаковой доли churn customer во **Франции** и **Испании**, их можно объединить в одну группу. 
А в этом случае нам уже будет достаточно **одного столбца** для категории `from_germany`


3. Было бы неплохо уточнить, что там произошло в Германии (и за какой период у нас данные). Возможно, не совсем корректно добавлять этот признак в предсказательную модель. Пока оставляем.

Решено:
- для обеих категорий cделаем `маппинг`


- перед этим попробуем **закодировать** страны 3-мя цифрами и посмотреть `корреляцию` (по аналогии с num_of_products).

```python
def ordinal_enc_geography(df):
    """ Кодирует категорию со странами для пайплайна"""
    data = df.copy()
    mapping = {"France": 0, "Germany": 1, "Spain": 2}
    data.loc[:, "geography"] = df["geography"].map(mapping)
    return data   
 
    
print('Корреляция географии клиентов и доли churn customer, Ordinal-Encoding:')
display(df_ml[['geography', 'exited']] 
        .pipe(ordinal_enc_geography)
        .corr(method='spearman')['exited']
        .drop(['exited']).to_frame()
        .sort_values(by='exited')
        .style.background_gradient(cmap='coolwarm'))
```

**Вывод 3**

Получается, `метод encoding` влияет на **интерпретацию корреляции**. Или нет?:)


- `OrdinalEncoding` указывает **на порядок**. Поэтому, если бы мы оставили все страны, то **для машинного обучения** корректнее было бы делать `OneHotEncoding` (в количестве продуктов порядок есть, а у стран в данном случае нет)


- Завершим  наши эксперименты  **маппингом**, чтобы теперь можно было сравнить корреляции внутри всех признаков. Мы получим два дополнительных признака для предсказательной модели: `from_germany` и `gender_id`.

редеет.

Таким образом, мы можем попробовать **разделить на категории** по границам:
  - 0 `баланс = 0`
  - 1 `баланс <= 50 000`
  - 2 `баланс < 119699` (среднее)
  - 3 `баланс < 200 000` 
  - 4 `баланс >= 200 000`

```python
# Функции для пайплайнов

def balance_categ(balance):
    '''Возвращает код для Ordinal Encoding баланса
    
    5 категорий
    '''    
    if balance == 0:
        return 0
    if balance <= 50000:
        return 1
    elif balance < 119699:
        return 2  
    elif balance < 200000:
        return 3   
    else:
        return 4  
    
def show_exited_corr(df, col_name):
    ''' Показывает корреляцию между балансом и другим признаком '''
    
    print(f"Корреляция между балансом клиентов и {col_name}:",
          f"{df['exited'].corr(df[col_name]): .3f}")
    return df
```

```python
# Добавляем все преобразования в пайплайн
# Пока экспериментируем, изменения не сохраняем
print("=> Категории для баланса с помощью Ordinal Encoding <=\n")
display(
    df_ml
    # Создадим дополнительный столбец для категорий баланса
    .assign(balance_cat=lambda x: x["balance"].apply(balance_categ))
    # Смотрим корреляцию категорий с целевым признаком
    .pipe(show_exited_corr, "balance_cat")
    # Считаем процент churn клиентов в категориях
    .pivot_table(index="balance_cat", values="exited")
)

print("\n Доля churn customer среди всех клиентов:")
df.exited.value_counts(normalize=True).to_frame()
```

После **добавления категорий** нам стало проще увидеть зависимости:

 1. **Нулевой баланс**. Здесь есть только `13%` churn customer. Возможно это свеженькие, новые клиенты, и они не успели принять решение уйти из банка.
 
 
 2. На втором месте - среднячок (2),(3). Все те, кто вошли в чашку виски) `24-25%`
 
 
 3. Самый большой процент сбежавших клиентов (как мы и увидели на графике scatter plot) среди тех, у кого:

 - слишком низкий баланс (1) < 50 000. `35%`
 - очень высокий (4) > 200 000. `56%`. 

Вот только **корреляция ухудшилась**:( Благими намерениями..

Попробуем закодировать **процентом churn customer** в каждой группе клиентов в зависимости от их баланса.

```python
def balance_categ_target(balance):
    '''Возвращает код для Target Encoding баланса
    
    5 категорий
    '''  
    if balance == 0:
        return 0.13
    elif balance <= 50000:
        return 0.36
    elif balance < 119699:
        return 0.24   
    elif balance < 200000:
        return 0.23   
    else:
        return 0.56  
    
print("=> Категории для баланса с помощью Target Encoding <=\n")
df_ml = (
    df_ml
    # Создадим дополнительный столбец для категорий баланса
    .assign(balance_cat_tg=lambda x: x["balance"].apply(balance_categ_target))
    # Смотрим корреляцию баланса с целевым признаком
    .pipe(show_exited_corr, "balance_cat_tg")
)
```

**Вывод 3**

- **Тип encoding** влияет на **корреляцию** и, скорее всего, на `качество обучения` моделей тоже (пока мы это не проверили)


- Закодировав группы клиентов **процентом churn customer**, мы смогли увеличить корреляцию параметра `balance` с `0.11 до 0.13`.
Другой вопрос, понравился бы такой финт предсказательной модели?


- `Target кодирование` нужно будет обязательно проводить **отдельно от тестовой выборки**! А пока результат encoding мы сохранили в столбце `balance_cat_tg` для матрицы корреляции.

3. Максимальное значение 850 - это клиенты, за которыми банки стоят в очереди:)

Создадим **категории кредитного скоринга** и посмотрим, как между ними распределятся churn customer). Категории возьмем из шкалы выше, но с учетом нашего распределения добавим для эксперимента **2 дополнительные**:

- `<= 400 баллов` экстремально низкий (аутлайеры)

- `= 850 балла` экстремально высокий (наш пик на графике)

```python
# Функции для пайплайна
def show_df(df):
    '''Выводит датафрейм'''
    display(df.head(5))
    return df


def credit_score_categ(score):
    '''Возращает категорию доя скоринга
    
    5 категорий
    '''
    if score <= 400:
        return '0 экстремально низкий'
    elif score < 600:
        return '1 очень низкий'   
    elif score < 650:
        return '2 низкий'    
    elif score < 690:
        return '3 высокий'  
    else:
        return '4 очень высокий'


# Добавляем в пайплайн преобразование скоринга в категории
# Пока экспериментируем, изменения не сохраняем
print("Доля churn customer среди клиентов с разным кредитным скорингом: ")
display(df
        # Создадим дополнительный столбец для категорий скоринга
        .assign(credit_score_cat=lambda x: x['credit_score']
                .apply(credit_score_categ))        
          
        # Считаем процент churn клиентов в категориях
        .pivot_table(index='credit_score_cat', values='exited'))

print("\n Доля churn customer среди всех клиентов:")
df.exited.value_counts(normalize=True).to_frame()
```

## Подготовка данных для машинного обучения

### Encoding категорий и  пайплайн  предобработки для исходного датасета

На этом этапе мы делаем дополнительные преобразования общего датасета, необходимые для **машинного обучения**.
1. **Изменения** сохраняем в новом датасете - `df_ml` и удаляем ненужные столбцы


2. Все шаги (включая предыдущие по предобработке данных) добавляем в **один общий пайплайн**


3. (!) Данные с пропусками в `tenure` оставляем, т.к. данный признак не будет участвовать в предсказаниях.


Исходный датасет оставляем - у нас к нему обязательно появятся вопросы:)
Далее подготовленный датасет `df_ml` можно будет разделить на обучающую и тестовую выборки.

```python
# Функции для пайплайна предобработки данных


def encode_gender(dfr):
    """Кодирует поле gender

    0 - Мужчина
    1 - Женщина
    """

    mapping = {"Male": 0, "Female": 1} 
    dfr["gender"] = dfr["gender"].map(mapping)

    return dfr


def encode_OHE(dfr):
    """One-Hot encoding для полей типа object"""

    # Используем one-hot encoding
    # удаляем столбец с dummy-ловушкой
    dfr = pd.get_dummies(dfr, drop_first=True)

    # get_dummies применяется ко всем столбцам типа object
    # данная функция идет последней в пайплайне
    return dfr


def columns_to_lower(dfr):
    """Переводит названия всех столбцов в нижний регистр"""

    dfr.columns = dfr.columns.str.lower()
    return dfr


def encode_geography(dfr):
    """Создает столбец from_germany

    Возвращает 1, если клиент из Германии
    Возвращает 0, если из Испании или Франции
    """
    dfr["from_germany"] = (dfr["geography"] == "Germany") * 1
    return dfr


def credit_score_cat_ord(score):
    """Возращает категорию доя скоринга

    3 категории
    """

    if score <= 400:
        return 1
    elif score < 650:
        return 2
    else:
        return 3
```

```python
# Создаем новый датафрейм для обучения моделей
# В общий пайплайн добавляем шаги, которые делали на предыдущих этапах
df_ml = (
    data.copy()
    # Удалим столбцы RowNumber, Surname и CustomerId
    .drop(["RowNumber", "Surname", "CustomerId"], axis=1)
    # Переименуем столбцы
    .rename(
        columns={
            "CreditScore": "credit_score",
            "NumOfProducts": "num_of_products",
            "HasCrCard": "has_cr_card",
            "IsActiveMember": "is_active_member",
            "EstimatedSalary": "est_salary",
        }
    )
    # Переведем все столбцы в нижний регистр
    .pipe(columns_to_lower)
    # Кодируем категории
    .pipe(encode_gender).pipe(encode_geography)
    # Создаем новый столбец для target encoding кредитного скоринга
    .assign(
        credit_score_cat_ord=lambda x: x["credit_score"].apply(
            credit_score_cat_ord)
    )
)


# Смотрим, что получили:)
display(df_ml.head(7))
df_ml.info()
```

```python
# Создаем и обучаем Dummy модель для оценки адекватности наших моделей

strategies = ['most_frequent', 'prior', 'stratified',  'uniform']
dummy_results = []

for strat in strategies:
    dc = DummyClassifier(strategy=strat, random_state=random_id)
    dc.fit(X_train, y_train) 
    result1 = f1_score(y_test, dc.predict(X_test)) 
    result2 = roc_auc_score(y_test, dc.predict(X_test)) 
    dummy_results.append({strat: [result1, result2]}) 

dummy_results   
```

**Вывод 1**

Мм,интересные результаты:) Рандомная модель `uniform` не так уж и плоха, а что можешь ты?)

- `roc_auc_score = 0.5` Все верно, это значит, что NO-Skill модель не умеет ранжировать классы 0 и 1. Поэтому кривая roc_auc - это диагональная прямая y(TPR) = x (FPR)


- `f1_score = 0` для нескольких моделей. Скорее всего, они плохо предсказывают наш **minority** класс.
Как это проверить? Посмотрим на все метрики подробнее с помощью `classification_report`

```python
for strat in strategies:
    dc = DummyClassifier(strategy=strat, random_state=random_id)
    dc.fit(X_train, y_train) 
    
    print(f"Dummy-классификатор со стратегией {strat.upper()}:")
    
    # Параметр zero_division, чтобы избежать warning из-за нулевых метрик
    print(classification_report(y_test, dc.predict(X_test), zero_division=0))
```

**Вывод 2**

### Downsampling & Upsampling

#### Вводные

Задача данного блока улучшить результаты, которые мы получили с помощью параметра WEIGHT!

Специально для проекта мы сами **написали функции** для Downsampling & Upsampling, а также различных экспериментов с параметрами данных методов. Это позволит лучше понять, как они работают))


- Данные техники применяются **только к обучающему датасету**, а валидационный остается с прежним дисбалансом классов. Поэтому в случае кросс-валидации нам потребовалось бы добавить downsampling&upsampling в пайплайн подготовки данных 


- Для готовых функций (как `StandardScaler`) это можно сделать, например, с помощью параметра `make_pipeline`. Написание собственного класса выходит за рамки этого проекта, поэтому создадим одну дополнительную **валидационную выборку**.


- Будем выполнять downsampling/upsampling **итеративно**, пока не добьемся `максимального улучшения качества модели` (здесь обязательно при контроле `overfitting`)

```python
# Создаем обучающую и валидационную выборку из датасета X_train
# Параметр stratify для сохранения четкого баланса в выборках
X_train_cv, X_valid, y_train_cv, y_valid = train_test_split(
    X_train, y_train,
    test_size=0.2, random_state=random_id,
    stratify=y_train) 


# train_test_split не создает копии
# Создаем копии вручную для использования трансформеров
X_train_cv = X_train_cv.copy()
X_valid = X_valid.copy()

# Проводим нормализацию данных (для Логистической регрессии)
scaler = StandardScaler()
X_train_cv[numeric_columns] = scaler.fit_transform(X_train_cv[numeric_columns])
X_valid[numeric_columns] = scaler.transform(X_valid[numeric_columns])


print(f"Распределение сhurn-клиентов в обучающей и валидационной выборках:")
display(
    y_train
    .value_counts(normalize=True).to_frame()
    .join(y_train_cv.value_counts(normalize=True), rsuffix='_new_train')
    .join(y_valid.value_counts(normalize=True), rsuffix='_valid')
    .join(y_test.value_counts(normalize=True), rsuffix='_test')
)

print("Размеры новых выборок:")
print(f"Обучающая -> {X_train_cv.shape[0]} |"
      + f"Валидационная -> {X_valid.shape[0]}")
```

### Построение ROC_AUC кривых

**Идеальной моделью** считается та, которая проходит через верхний левый угол:
